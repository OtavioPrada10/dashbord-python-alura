**Contagem de quantos filmes por linguagem de origem**

```python
#Contagem de quantos filmes por linguagem de origem
contagem_de_lingua =tmdb['original_language'].value_counts().to_frame().reset_index()
contagem_de_lingua.columns = ['lingua','total']
contagem_de_lingua.head()
```

```python
sns.barplot(data = contagem_de_lingua, x = 'lingua', y = 'total')
```

```python
sns.countplot(data=tmdb, x='original_language')
```

```python

contagem_de_lingua.plot(kind='pie', y="total", labels=contagem_de_lingua['lingua'])
```

```python
total_por_lingua = tmdb["original_language"].value_counts()
total_geral  = total_por_lingua.sum()
total_de_ingles = total_por_lingua['en']
total_do_resto = total_geral - total_de_ingles
print(total_geral, total_de_ingles, total_do_resto)
```

```python
dados = {
"lingua": ["ingles", "outros"],
"totlal": [total_do_resto, total_de_ingles]
}

dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(data=dados, y= dados['totlal'], x=dados['lingua'])
```

```python
dados.plot(kind='pie', y="totlal", labels=dados['lingua'])
```
